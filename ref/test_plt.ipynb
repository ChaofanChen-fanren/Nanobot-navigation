```python
import numpy as np
import matplotlib.pyplot as plt

# 初始位置
x0 = 430
y0 = 572

# 创建一个从 0 到 2*pi 的向量 t
t = np.linspace(0, 200, 200)

# 计算 sin(t)
x = x0 + t
y = y0 + 200*np.sin(t/50)

# 画图
plt.figure()
plt.plot(x, y)
plt.title('Desired trajectory plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
